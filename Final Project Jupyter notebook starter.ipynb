```python
#drop all of the rows where the manufacturer is missing --> need to figure out how to do this

df = df[~df['manufacturer'].isna()]
df
```

```python
#create a regional data frame for further analysis

regional_df = df[['region','price', 'model', 'size', 'condition', 'state']]
regional_df
```

```python
#replace missing condition with the word 'unknown'

regional_df['condition'].fillna(value='unknown', inplace=True)
regional_df
```

```python
#the (count of) condition of the cars
regional_df['condition'].value_counts()
```

```python
#replace missing model values with the word 'unknown'
regional_df['model'].fillna(value='unknown', inplace=True)
regional_df
```

```python
#replace missing size values with the word 'unknown'
regional_df['size'].fillna(value='unknown', inplace=True)
regional_df
```

```python
regional_df['model'].value_counts()
```

```python
#replace missing size with the word unknown
regional_df['size'].value_counts()
```

```python
#drop all the rows where the model, size, and condition are all unknown

regional_df = regional_df[~(regional_df['condition']=='unknown') | ~(regional_df['model']=='unknown') | ~(regional_df['size']=='unknown')]
regional_df
```

```python
#set the national average at $35,000

nat_avg = 35000
```

```python
#getting the mean of price before cleansing
regional_df['price'].mean()
```

```python
#getting the number of cars that have a price of 0
regional_df[regional_df['price']==0]
```

```python
#find all the rows where price is above $300k

regional_df[regional_df['price']>300000]
```

```python
regional_df = regional_df[~(regional_df['price']>300000)]
regional_df
```

```python
#replace all the prices that are 0 with the national average
regional_df['price'].replace(0, 35000)
```

```python
#taking the average after the price swap

regional_df['price'].mean()
regional_df['price'].std()
```

```python
#getting the average price by state
regional_df.groupby('state')['price'].mean()
```

```python
#load in the state table

state_df = pd.read_csv('states.csv')

regional_df = regional_df.merge(state_df, on='state')
regional_df
```

```python
#cleaned and ready for analysis
regional_df
```

```python
regional_df.to_csv('clean_data.csv')
```

```python
clean_df = pd.read_csv('clean_data.csv')
clean_df
```

```python
plt.boxplot(clean_df['price'], vert=False)
plt.xlabel('Price')
plt.title('Barplot of Price');
```

```python
#getting total/average price by region
national_sum = clean_df.groupby('national_region')['price'].sum()
national_avg = clean_df.groupby('national_region')['price'].mean()
```

```python
#national sum by region
plt.bar(x=national_sum.index, height=national_sum)
plt.title('Sum of Price by Region (in $1000s)');
```

```python
#national avg by region
plt.bar(x=national_avg.index, height=national_avg)
plt.title('Average of Price by Region');
```

```python
#getting the top 10 most popular models (by count)
top_three_mod = clean_df['model'].value_counts().head(10)
plt.bar(top_three_mod.index, height=top_three_mod)
plt.xticks(rotation=45)
plt.title('Top 10 Most Popular Models by Frequency');
```

```python
plt.pie(top_three_mod, labels = top_three_mod.index);
```

```python
size_list = clean_df['size'].value_counts()
size_list = size_list[1:]
```

```python
plt.bar(x=size_list.index, height=size_list)
plt.title('Frequency of Car Sizes');
```

```python
#getting the conditions without unknown
cond_list = clean_df['condition'].value_counts()
cond_list = cond_list[1:]
cond_list
```

```python
plt.bar(x=cond_list.index, height=cond_list)
plt.title('Frequency of Car Condtions');
```
